# 그래프 구조 변경

> - `Figure, Axes, Axis`로 구성
>> - `Figure`: [그림이 그려지는 캔버스나 종이](https://matplotlib.org/stable/api/figure_api.html)
>> - `Axes`: [하나의 그림](https://matplotlib.org/stable/api/axes_api.html)
>> - `Axis`: [그림의 X축이나 Y축 등의 축을 의미](https://matplotlib.org/stable/api/axis_api.html)
> - `Figure`는 1개 이상의 `Axes`를 포함하고, Axes는 다시 2개 이상의 `Axis` 포함
> - 일반적인 plot 명령 등을 실행하면 자동으로 Figure를 생성해주기 때문에 별도로 `Figure, Axes, Axis` 생성 명령어를 사용하진 않아도 됨
> - 세부 구조를 변경/응용 하고 싶은 경우 `Figure, Axes, Axis` 사용하여 반영
>

```python
# Figure & Axes의 관계
# subplot(행갯수, 열갯수, 행열값들 중 인덱스)
plt.figure(figsize=(10, 5))    # Figure 생성
axes1 = plt.subplot(1,2,1)    # 1행 2열의 Figure에서 1열 Axes를 의미
axes2 = plt.subplot(1,2,2)    # 1행 2열의 Figure에서 2열 Axes를 의미
```

```python
# Figure & Axes의 관계
# subplot(행갯수, 열갯수, 행열값들 중 인덱스)
# 각 subplot에 맞게 그림 그리기
plt.figure(figsize=(10, 5))    # Figure 생성

x = [1, 2, 3]
x_second = [-3, -1, 1, 3, 5]
y = [10, 20, 30]
y_second = [15, 20, 25, 30, 35]  

plt.subplot(1,2,1)    # Axes 1개 생성
plt.plot(x, y, label='first values', linestyle='--',
           linewidth=5, marker='*', markersize=15, 
           markeredgecolor='red', markeredgewidth='2', markerfacecolor='yellow') 
plt.title('테스트 그림1')
plt.legend(loc='lower right')    
plt.xlabel('Time1')    
plt.ylabel('Sales1') 

plt.subplot(1,2,2)    # Axes 1개 생성
plt.plot(x_second, y_second, label='second values', color='red')
plt.title('테스트 그림2')
plt.legend(loc='lower right')    
plt.xlabel('Time2')    
plt.ylabel('Sales2') 

plt.show()
```

```python

plt.figure(figsize=(10, 5))   

x = [1, 2, 3]
x_second = [-3, -1, 1, 3, 5]
y = [10, 20, 30]
y_second = [150, 200, 250, 300, 350]  

plt.plot(x, y, label='first values', linestyle='--',
         linewidth=5, marker='*', markersize=15, 
         markeredgecolor='red', markeredgewidth='2', markerfacecolor='yellow') 
plt.plot(x_second, y_second, label='second values', color='red')
plt.title('테스트 그림')
plt.legend(loc='lower right')    
plt.xlabel('Time')    
plt.ylabel('Sales')   
plt.xlim(-5, 7)    # 변경
plt.ylim(0, 350)    # 변경
plt.grid(axis='y')  
plt.show()
```

```python
# twinx 함수로 axes를 생성하여 새로운 축으로 그림 생성
fig, axes = plt.subplots(figsize=(10,5))    # 갯수를 모르는 임의 그림 생성
axes_another = axes.twinx()    # 신규 axes 생성

x = [1, 2, 3]
x_second = [-3, -1, 1, 3, 5]
y = [10, 20, 30]
y_second = [150, 200, 250, 300, 350] 
axes.set_title('테스트 그림1')
axes.set_xlabel('Time1')    

# axes 1
axes.plot(x, y, label='first values', linestyle='--',
          linewidth=5, marker='*', markersize=15, 
          markeredgecolor='red', markeredgewidth='2', markerfacecolor='yellow') 

axes.legend(loc='upper left')    
axes.set_ylabel('Sales1') 

# axes 2
axes_another.plot(x_second, y_second, label='second values', color='red')
axes_another.legend(loc='lower right')    
axes_another.set_ylabel('Sales2') 

plt.show()
```

## Pie Chart

### 기본

```python
# Pie chart 기본
import matplotlib as mpl
import matplotlib.pylab as plt

sizes = [15, 30, 45, 10]

plt.pie(sizes)
plt.show()
```

### labels, colors 속성

```python
# Pie chart 기본
# 표시 레이블과 색상을 원하는 대로 변경 가능
sizes = [15, 30, 45, 15, 5]
labels = ['월', '화', '수', '목', '금']
colors = ['green', 'gold', 'lightskyblue', 'lightcoral', 'yellow']

plt.figure(figsize=(10,5))
plt.pie(sizes, labels=labels, colors=colors) # 
plt.title('스트레스 정도')
plt.show()
```

### shadow, startangle, autopct 속성

```python
# Pie chart 기본
# 그림자 정도, 시작위치, 값의 표시 가능
sizes = [15, 30, 45, 15, 5]
labels = ['월', '화', '수', '목', '금']
colors = ['green', 'gold', 'lightskyblue', 'lightcoral', 'yellow']

plt.figure(figsize=(10,5))
plt.pie(sizes, labels=labels, colors=colors, 
        shadow=True, startangle=90, autopct='%.1f%%') # 
plt.title('스트레스 정도')
plt.show()
```

### explode 속성

```python
# Pie chart 기본
# 값의 위치 강조 표시 가능
sizes = [15, 30, 45, 15, 5]
labels = ['월', '화', '수', '목', '금']
colors = ['green', 'gold', 'lightskyblue', 'lightcoral', 'yellow']
explode = (0, 0.2, 0.3, 0, 0)

plt.figure(figsize=(10,5))
plt.pie(sizes, labels=labels, colors=colors, 
        shadow=True, startangle=90, autopct='%.1f%%',
        explode=explode)
plt.title('스트레스 정도')
plt.show()
```

## Bar Chart

### 기본

```python
# Bar chart 기본
import matplotlib as mpl
import matplotlib.pylab as plt

x = [1,2,3,4,5]
y = [9,5,3,7,5]

plt.bar(x, y)
plt.show()
```

### xerr, yerr 속성

오차 정도 표현

```python
# Bar chart 꾸미기
# bar 함수 내부 인자 xerr 또는 yerr 사용하여 오차 정도를 표현 가능
x = [1,2,3,4,5]
y = [9,5,3,7,5]

plt.figure(figsize=(10,5))
plt.bar(x, y, yerr=[1,2,2,0.5,1])  # 
plt.title('Bar Chart')
plt.xlabel('요일')
plt.ylabel('매출액')
plt.xticks(ticks=x, labels=['월', '화', '수', '목', '금'])
plt.grid(axis='y')
plt.show()
```

### alpha 속성

```python
# Bar chart 꾸미기
# bar 함수 내부 인자 alpha는 그래프의 투명도 (0:투명, 1:불투명)
x = [1,2,3,4,5]
y = [9,5,3,7,5]

plt.figure(figsize=(10,5))
plt.bar(x, y, yerr=[1,2,2,0.5,1], alpha=0.3) # 0 ~ 1
plt.title('Bar Chart')
plt.xlabel('요일')
plt.ylabel('매출액')
plt.xticks(ticks=x, labels=['월', '화', '수', '목', '금'])
plt.grid(axis='y')
plt.show()
```

## Histogram

### 기본

```python
# Histogram plot 기본
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np

data = np.random.normal(50, 15, size=1000)

plt.hist(data)
plt.show()
```

### color, alpha, edgecolor, bins 속성

```python
# Histogram plot 기본
# 색상, 투명도, 엣지색상, 구간 갯수 설정 가능
data = np.random.normal(50, 15, size=1000)

plt.figure(figsize=(10,5))
plt.hist(data, color='blue', alpha=0.3, edgecolor='red', bins=30)
plt.title('Histogram')
plt.xlabel('연령')
plt.ylabel('구간 당 COVID 인원수')
plt.show()
```

### cumulative 속성

누적 히스토그램

```python
# Histogram plot 기본
# 누적 히스토그램
data = np.random.normal(50, 15, size=1000)

plt.figure(figsize=(10,5))
plt.hist(data, color='red', alpha=0.3, edgecolor='black', bins=30, cumulative=True)
plt.title('Histogram')
plt.xlabel('연령 표준화')
plt.ylabel('구간 당 COVID 인원수')
plt.show()
```

```python
# Histogram plot 기본
# 여러 데이터 중복
data = np.random.normal(50, 15, size=1000)
data2 = np.random.normal(50, 15, size=1000) * 0.5

plt.figure(figsize=(10,5))
plt.hist(data, color='red', alpha=0.3, edgecolor='black', bins=30, cumulative=True)
plt.hist(data2, color='blue', alpha=0.3, edgecolor='black', bins=30, cumulative=True)
plt.title('Histogram')
plt.xlabel('연령 표준화')
plt.ylabel('구간 당 COVID 인원수')
plt.show()
```

## Box Plot



- **Maximum:** `Q3에서 1.5 * Interquartile Range(IQR, Q3-Q1)` 위 범위까지 중 최상위
- **Quantile 25%(Q3):** 상위 25%
- **Quantile 50%(Q2, Median):** 중앙값
- **Quantile 25%(Q1):** 등수 상 하위 25%
- **Minimum:** `Q1에서 1.5 * Interquartile Range(IQR, Q3-Q1)` 아래 범위까지 중 최하위


### 기본

```python
# Box plot 기본
import matplotlib as mpl
import matplotlib.pylab as plt

data = np.random.normal(50, 15, size=1000)

plt.boxplot(data)
plt.show()
```

### norch 속성

```python
# Box plot 기본
# 제목, 라벨, 중앙값의 표현 가능
data = np.random.normal(50, 15, size=1000)

plt.figure(figsize=(10,5))
plt.boxplot(data, notch=True)
plt.title('Box Plot')
plt.ylabel('Distribution')
plt.show()
```

### 기본

```python
# Violin plot 기본
import matplotlib as mpl
import matplotlib.pylab as plt

data = np.random.normal(50, 15, size=1000)

plt.violinplot(data)
plt.show()
```

### quantiles 속성

```python
# Box plot 기본
# 4분위 Quantile 표현 가능
data = np.random.normal(50, 15, size=1000)

plt.figure(figsize=(10,5))
plt.violinplot(data, quantiles=[0.25, 0.5, 0.75])
plt.title('Violin Plot')
plt.ylabel('Distribution')
plt.show()
```

## Scatter Plot
- `2차원의 데이터 표현 및 관계`를 살펴보는데 많이 사용    
- 특히 `실수인 데이터`의 관계성을 한눈에 보는데 효과   

### 기본

```python
# Scatter plot 기본
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np

x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
print(x)
print(y)

plt.scatter(x, y)
plt.show()
```

### label, alpha 속성

```python
# Scatter plot 기본
# 여러 옵션들
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0

plt.figure(figsize=(10,5))
plt.scatter(x, y, label='COVID', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('발생경과일')
plt.ylabel('확진자수')
plt.legend(loc='upper left')
plt.show()
```

### c, cmap 속성

```python
# Scatter plot 기본
# 값별 색상부여 가능
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
colors = np.arange(len(x))

plt.figure(figsize=(10,5))
plt.scatter(x, y, label='COVID', c=colors, alpha=0.5, cmap='hsv')
plt.colorbar()
plt.title('Scatter Plot')
plt.xlabel('발생경과일')
plt.ylabel('확진자수')
plt.legend(loc='upper left')
plt.show()
```

### Stem and Leaft Plot

## Surface Plot 

- 3차원 전용 `Axes`를 추가하여 입체적으로 표현   
- 3차원에 대응되는 2차원 입력값 `x, y`는 `meshgrid`로 포인트별 행렬 변환 필수

```python
# Surface 기본
# meshgrid 함수로 x, y 데이터를 필수로 묶음
# meshgrid 함수는 인수로 반드시 array를 입력받음
import matplotlib as mpl
import matplotlib.pylab as plt

x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig, axes = plt.subplots(figsize=(10,10), subplot_kw={'projection':'3d'})  
axes.plot_surface(x, y, z)
plt.show()
```

### cmap, set_title

```python
# Surface 기본
# 색상, 제목 변경 가능
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig, axes = plt.subplots(figsize=(10,10), subplot_kw={'projection':'3d'})  
axes.plot_surface(x, y, z, cmap='hot')
axes.set_title('3D Example')
plt.show()
```

### colorbar(shrink, aspect)

```python
# Surface 기본
# legend bar 추가
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig, axes = plt.subplots(figsize=(10,10), subplot_kw={'projection':'3d'})  
surf = axes.plot_surface(x, y, z, cmap=plt.cm.coolwarm)
fig.colorbar(surf, shrink=0.5, aspect=5)
axes.set_title('3D Example')
plt.show()
```

## Contour Plot 
- 3차원 데이터를 명암이 아닌 `등고선`을 사용하여 표현    
- 3차원에 대응되는 2차원 입력값 `x, y`는 `meshgrid`로 포인트별 행렬 변환 필수

### 기본

```python
import matplotlib as mpl
import matplotlib.pylab as plt

x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10,5))
plt.contour(x, y, z)
plt.show()
```

### colors, title

```python
# Contour 기본
# 제목, 선색상 변경 가능
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10,5))
plt.contour(x, y, z, colors='black')
plt.title('Contour Plot')
plt.show()
```

### contourf

```python
# contourf 함수로 색깔도 반영 가능
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10,5))
plt.contourf(x, y, z)
plt.title('Contour Plot')
plt.show()
```

### contourf(alpha, cmap)

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10,5))
plt.contourf(x, y, z, alpha=0.75, cmap='jet')
plt.title('Contour Plot')
plt.show()
```

```python
# Contour 기본
# contourf & contour 함수 중복 가능 가능
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10,5))
plt.contour(x, y, z, colors='black')
plt.contourf(x, y, z, alpha=0.75, cmap='jet')
plt.title('Contour Plot')
plt.show()
```
